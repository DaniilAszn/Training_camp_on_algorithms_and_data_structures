## 2.1.1. [1672. Richest Customer Wealth](https://leetcode.com/problems/richest-customer-wealth/description/)

1

```python
class Solution:
    def maximumWealth(self, accounts: List[List[int]]) -> int:
        ans = 0

        for acc in accounts:
            cur_sum_acc = sum(acc)
            if cur_sum_acc > ans:
                ans = cur_sum_acc

        return ans
```

## 2.1.2. [1822. Sign of the Product of an Array](https://leetcode.com/problems/sign-of-the-product-of-an-array/description/)

1

```python
class Solution:
    def arraySign(self, nums: List[int]) -> int:
        ans = 1

        for num in nums:
            if num == 0:
                return 0
            ans *= num

        return 1 if ans > 0 else -1
```

## 2.1.3. [1431. Kids With the Greatest Number of Candies](https://leetcode.com/problems/kids-with-the-greatest-number-of-candies/description/)

1

```python
class Solution:
    def kidsWithCandies(self, candies: List[int], extraCandies: int) -> List[bool]:
        max_sweet = max(candies)
        ans = []

        for c in candies:
            ans.append(c + extraCandies >= max_sweet)

        return ans
```

## 2.1.4. [1470. Shuffle the Array](https://leetcode.com/problems/shuffle-the-array/description/)

1

```python
class Solution:
    def shuffle(self, nums: List[int], n: int) -> List[int]:
        ans = []

        for i in range(n):
            ans.append(nums[i])
            ans.append(nums[i + n])

        return ans
```

## 2.1.5. [1491. Average Salary Excluding the Minimum and Maximum Salary](https://leetcode.com/problems/average-salary-excluding-the-minimum-and-maximum-salary/description/)

1

```python
class Solution:
    def average(self, salary: List[int]) -> float:
        max_s = -float('inf')
        min_s = float('inf')
        sum_s = 0
        len_s = 0

        for s in salary:
            max_s = max(max_s, s)
            min_s = min(min_s, s)
            sum_s += s
            len_s += 1

        return (sum_s - max_s - min_s) / (len_s - 2)
```

## 2.1.6. [605. Can Place Flowers](https://leetcode.com/problems/can-place-flowers/description/)

2

```python
class Solution:
    def canPlaceFlowers(self, flowerbed: List[int], n: int) -> bool:
        flowerbed = [0] + flowerbed + [0]

        for i in range(1, len(flowerbed) - 1):
            if (flowerbed[i - 1] == 0) and (flowerbed[i] == 0) and (flowerbed[i + 1] == 0):
                flowerbed[i] = 1
                n -= 1

        return n <= 0
```

## 2.1.7. [1313. Decompress Run-Length Encoded List](https://leetcode.com/problems/decompress-run-length-encoded-list/description/)

1

```python
class Solution:
    def decompressRLElist(self, nums: List[int]) -> List[int]:
        ans = []

        for i in range(0, len(nums) - 1, 2):
            ans += [nums[i + 1]] * nums[i]

        return ans
```

## 2.1.8. [1299. Replace Elements with Greatest Element on Right Side](https://leetcode.com/problems/replace-elements-with-greatest-element-on-right-side/description/)

4

```python
class Solution:
    def replaceElements(self, arr: List[int]) -> List[int]:
        max_elem = -1

        for i in range(len(arr) - 1, -1, -1):
            temp_elem = arr[i]
            arr[i] = max_elem
            max_elem = max(max_elem, temp_elem)

        return arr
```

## 2.1.9. [1464. Maximum Product of Two Elements in an Array](https://leetcode.com/problems/maximum-product-of-two-elements-in-an-array/description/)

2

```python
# solution 1
class Solution:
    def maxProduct(self, nums: List[int]) -> int:
        max_1 = -float('inf')
        max_2 = -float('inf')

        for num in nums:
            if num > max_1:
                max_2 = max_1
                max_1 = num
            elif num > max_2:
                max_2 = num

        return (max_1 - 1) * (max_2 - 1)

# solution 2
class Solution:
    def maxProduct(self, nums: List[int]) -> int:
        ans = -1
        max_prev = nums[0]

        for i in range(1, len(nums)):
            ans = max(ans, (nums[i] - 1 ) * (max_prev - 1))
            max_prev = max(max_prev, nums[i])

        return ans
```

## 2.1.10. [2574. Left and Right Sum Differences](https://leetcode.com/problems/left-and-right-sum-differences/description/)

4

```python
class Solution:
    def leftRightDifference(self, nums: List[int]) -> List[int]:
        ans = []
        left_sum = 0
        right_sum = sum(nums)

        for num in nums:
            ans.append(abs(left_sum - (right_sum - num)))
            left_sum += num
            right_sum -= num

        return ans
```

## 2.1.11. [724. Find Pivot Index](https://leetcode.com/problems/find-pivot-index/description/)

3

```python
class Solution:
    def pivotIndex(self, nums: List[int]) -> int:
        left_sum = 0
        right_sum = sum(nums)

        for i in range(len(nums)):
            if left_sum == (right_sum - nums[i]):
                return i
            left_sum += nums[i]
            right_sum -= nums[i]

        return -1
```
